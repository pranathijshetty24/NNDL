```python
import numpy as np
x = np.array([5, 10, 15])
w = np.array([[0.3, 0.4], [0.5, 0.6], [0.7, 0.8]])
b = np.array([0.2, 0.3])
z = np.dot(x,w)+b
print("output",z)
def sigmoid(x):
  return 1/(1+np.exp(-x))
a = sigmoid(z)
print("activated output:",a)
w2 = np.array([[0.08, 0.15]])
b2 = np.array([0.2])
z2 = np.dot(a,w2.T)+b2
a2 = sigmoid(z2)
print("final output:",a2)
y = 0.75
loss = (y-a2)**2
print("loss:",loss)
lr = 0.005
dloss_da2 = -2*(y-a2)
da2_dz2 = a2*(1-a2)
dz2_dw2 = a
grad_w2 = dloss_da2*da2_dz2*dz2_dw2
w2 = w2 -lr* grad_w2.reshape(2,1)
print("updtaed w2:",w2)
```
